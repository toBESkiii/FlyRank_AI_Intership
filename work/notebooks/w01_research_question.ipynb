## 1. My lane and why

My provisional lane is **Refresh / Content Opportunity Scoring**.

I selected this lane because a website may contain thousands of pages, making it difficult for an SEO specialist or content editor to review every page manually. This project will investigate whether search-performance, content and engagement signals can help identify which pages should be reviewed first.

The intended output will be a ranked review queue containing priority scores, suggested actions and clear reasons for each recommendation. The system will support human decision-making rather than automatically deciding that a page must be changed.


```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import pandas as pd

DATA_URL = (
    "https://raw.githubusercontent.com/"
    "toBESkiii/FlyRank_AI_Intership/main/"
    "data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(DATA_URL)

print("Starter dataset loaded successfully.")
print("Dataset shape:", df.shape)
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

## 2. The question: decision, action, cost of a wrong call

### Research question

**Which website pages should an SEO specialist or content editor review first for refresh, expansion, protection, pruning or monitoring, based on available search-performance, content and engagement signals?**

The unit of analysis is one website page.

The decision being improved is which pages should receive limited editorial attention first. The intended output is a ranked review queue showing the priority of each page, a possible action and the reason for the recommendation.

An SEO specialist or content editor could use the output to investigate the highest-priority pages and decide whether they should be refreshed, expanded, protected, consolidated, pruned or monitored.

A false positive could waste time by recommending a page that does not require attention. A false negative could cause the team to overlook an important page that is losing visibility or has a valuable improvement opportunity.

Data or machine learning may help because page performance can depend on several signals working together. However, the purpose of the project is not simply to train a model. Its purpose is to improve how pages are prioritised for human review.


```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

# Select pages currently labelled as declining
declining_pages = df[df["trend_direction"] == "down"]

# From those pages, keep pages with at least 100 impressions
declining_with_visibility = declining_pages[
    declining_pages["impressions_90d"] >= 100
]

print("Pages labelled as declining:", len(declining_pages))
print(
    "Declining pages with at least 100 impressions:",
    len(declining_with_visibility)
)
```

## 3. Quick look at the data (2-3 real numbers)

## 3. Quick look at the data

The starter dataset contains **30,000 website pages**.

Of these pages, **16,262 are labelled as declining**, representing approximately **54.2% of the dataset**. In addition, **13,152 pages are both declining and have received at least 100 impressions during the 90-day period**.

These figures show that the dataset contains thousands of possible review candidates. It would be difficult and time-consuming for an SEO specialist or content editor to inspect every declining page manually.

This supports the Refresh / Content Opportunity Scoring lane because a ranked review queue could help the team focus first on pages with stronger evidence of a meaningful opportunity. However, these numbers do not prove that every declining page needs to be refreshed.


```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Calculate the total number of pages
total_pages = len(df)

# Count the pages labelled as declining
declining_count = len(declining_pages)

# Count declining pages with at least 100 impressions
visible_declining_count = len(declining_with_visibility)

# Calculate percentages
declining_percentage = declining_count / total_pages * 100
visible_declining_percentage = visible_declining_count / total_pages * 100

print(f"Total pages: {total_pages:,}")
print(
    f"Pages labelled as declining: "
    f"{declining_count:,} ({declining_percentage:.1f}%)"
)
print(
    f"Declining pages with at least 100 impressions: "
    f"{visible_declining_count:,} "
    f"({visible_declining_percentage:.1f}%)"
)
```

## 4. Careful words: what I can and cannot claim

This project will provide **observed, directional and decision-support evidence**.

It may show that certain combinations of search-performance, content and engagement signals are associated with pages receiving higher review-priority scores. It may also show that one prioritisation method performs better than a simple rule-based baseline within the available dataset.

The output may help an SEO specialist or content editor decide which pages deserve investigation first. However, a high-priority score will not automatically mean that a page must be changed.

This project will not claim that it has discovered or can predict Google’s ranking algorithm. It will not prove that a particular feature caused a page to grow or decline. It will also not guarantee that refreshing, expanding or changing a recommended page will improve its future search performance.

A page appearing high in the ranked review queue will only mean that the available data provides stronger reasons to investigate that page before lower-ranked pages.
